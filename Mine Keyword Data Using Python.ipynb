```python
import pytrends 
from pytrends.request import TrendReq
import pandas as pd
import time
import datetime
from datetime import datetime, date, time
```

Play with the pytrends pakcage first

```python
ptrend = TrendReq() 
ptrend.build_payload(kw_list=['masks', 'printers', '','remdesivir', 'disinfecting wipes'], timeframe='today 12-m', geo = 'US', gprop='')
interest_df = ptrend.interest_over_time() 
print(interest_df.head())
```

```python
interest_df
```

Use pytrend.interest_over_time() function to check interest over time 

Draw a graph for the trends

```python
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set(color_codes=True)
ax = interest_df.plot(kind='line', figsize = (9,6), title = "Interest Over Time")
ax.set_xlabel('Date')
ax.set_ylabel('Trends Index')
ax.tick_params(axis='both', which='major', labelsize=13) # Used to adjust the sizes of ticks. Ticks examples are 'Aug' and '100' on x and y axis. 

plt.show()
```

### Insights: 
1. The interests of masks increased significantly at the end of march and had a dreasing trend in late May. In July the interests of masks is jumping back.
2. The interests of Remdesivir started to increase from March and the strongest interestes appeared in the middle of April. Interests have been decreasing since May.
3. Interests of printers didn't show significant changes. However, according to my observation, it is more difficult to purchase printers online since a couple of months ago. I need to segment the googel trend data for printers further.

### I am wondering what is the trend of printers in google shopping results

```python
printers_trend = TrendReq()
printers_trend.build_payload(kw_list=['printers'], timeframe='today 12-m', geo = 'US', gprop='froogle') #'froogle' will return just google shopping data
printers_int_df = printers_trend.interest_over_time() 
print(printers_int_df.head())
```

```python
ax = printers_int_df.plot(kind='line', figsize = (9,6), title = "Interest Over Time")
ax.set_xlabel('Date')
ax.set_ylabel('Trends Index')
ax.tick_params(axis='both', which='major', labelsize=13) # Used to adjust the sizes of ticks. Ticks examples are 'Aug' and '100' on x and y axis. 

plt.show()
```

### Insights: 

In general, interests of printers on Google Shopping have been increasing since February, which is consistent with my observation.
